1. Escolha uma palavra aleatória (ou deixe outro jogador escolher a palavra).
2. Inicialize as variáveis:
   - palavra_oculta: a palavra a ser adivinhada
   - palavra_mostrada: uma string vazia com o mesmo comprimento da palavra_oculta, para exibir as letras adivinhadas
   - tentativas_maximas: o número máximo de tentativas permitidas
   - tentativas: o número atual de tentativas (começa em 0)
   - letras_erradas: uma lista vazia para acompanhar as letras erradas
3. Enquanto tentativas < tentativas_maximas:
   4. Mostrar a palavra_mostrada e as letras erradas.
   5. Solicitar ao jogador uma letra ou adivinhação da palavra completa.
   6. Se a entrada for uma letra válida:
      7. Se a letra estiver na palavra_oculta:
         8. Atualize palavra_mostrada com a letra nas posições corretas.
         9. Se a palavra_oculta é igual a palavra_mostrada, o jogador venceu.
      10. Caso contrário, adicione a letra às letras_erradas.
      11. Incrementar tentativas.
   12. Senão, se a entrada for adivinhar a palavra completa:
      13. Verificar se adivinhação é igual à palavra_oculta.
      14. Se for igual, o jogador venceu.
      15. Caso contrário, incrementar tentativas.
16. Se tentativas atingirem tentativas_maximas, o jogador perdeu.
17. Exibir a palavra_oculta.
18. Perguntar ao jogador se deseja jogar novamente.


```python
import random


def jogo_da_forca():
    print('\nBem vindo ao jogo da forca!')
    print("Advinhe qual a palavra abaixo\n")
    
    #lista de palavras:
    palavras = ['Gustavo', 'Sattyla', 'Goiania', 'Campinas', 'Liquidificador', 'Beijo', 'Paralelepipedo', 'Umidificador', 'Geladeira', 'Caneca', 'Microondas']
    
    #escolhe aleatoriamente a palavra de uma lista
    escolha_aleatoria = random.choice(palavras)


    espaco = ['_' for letra in escolha_aleatoria]
    
    # numero de chances
    chances = 6
    
    #lista para as letras erradas
    letras_erradas = []
    
    while chances > 0:
        print(" ".join(espaco))
        print('\nChances Restantes:', chances)
        print('Letras erradas:', " ".join(letras_erradas))
        
        tentativa = input('\nDigite uma letra: ').lower()
        
        if tentativa in escolha_aleatoria.lower():
            index = 0
            for letra in escolha_aleatoria.lower():
                if tentativa == letra:
                    espaco[index] = letra
                index += 1
        else:
            chances -= 1
            letras_erradas.append(tentativa)
    
    if '_' in espaco:
        print('\n Você perdeu, a palavra era', escolha_aleatoria)
    
jogo_da_forca()
```
