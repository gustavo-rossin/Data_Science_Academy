```python
import numpy as np

n = np.random.randn(100000)

fig, axes = plt.subplots(1, 2, figsize= (10,4))

#plot 1:
axes[0].hist(n)
axes[0].set_title('histograma padrão')
axes[0].set_xlim(min(n), max(n))

#plot 2:
axes[1].hist(n, cumulative = True, bins = 50)
axes[1].set_title('histograma cumulativo')
axes[1].set_xlim(min(n), max(n))
```

```python
#Valores aleatórios: "entrevista com N pessoas para ver se eles são fumantes, qual sua altura e qual sua idade."

np.random.seed(42)
n = 1500
pct_smokers = 0.2

#Variáveis

flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

#df

dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante})

dados['flag_fumante'] = dados.flag_fumante.map({True: 'Fumante', False: 'Não Fumante'})
```

```python
dados.shape
```

```python
dados.head()
```

```python
sns.set(style = 'ticks')

#lmplot

sns.lmplot(
    data = dados,
    x = 'altura',
    y = 'peso',
    hue = 'flag_fumante',
    palette = ['tab:blue', 'tab:orange'],
    height = 7
)

plt.xlabel('altura (cm)')
plt.ylabel('peso (kg)')
plt.title('Relação entre peso e altura de fumantes e não fumantes')

sns.despine()

plt.show()
```
